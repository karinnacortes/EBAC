####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
# Base de dados para função do Exercício 1

pvl_br = pd.DataFrame(requests.get('https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl').json()['items'])
pvl_br.head(3)
````

````python
#1) Seu código aqui

def verifica_uf_valida(uf):
    c = 0
    for i in range(0,27):
        if uf == list(pvl_br['uf'].unique())[i] :
            c += 1
    if c == 1:
        return True
    else:
        return False
        
        
def verifica_tipo_interessado_valido(tipo_interessado):
    c = 0
    for i in range(0,2):
        if tipo_interessado == list(pvl_br['tipo_interessado'].unique())[i] :
            c += 1
    if c == 1:
        return True
    else:
        return False

def monta_pvl_uf(uf, tipo_interessado:str):
    
    if verifica_uf_valida(uf) and verifica_tipo_interessado_valido(tipo_interessado) :        
        df_uf = pd.DataFrame( (pvl_br[pvl_br['uf']== uf ]))
        return df_uf[df_uf['tipo_interessado']== tipo_interessado ]
    else:
        return print('Consulta não diponível com os parâmetros informados.\n\n',
                     'Valores válidos para "tipo_interessado": \n',list(pvl_br['tipo_interessado'].unique()),'\n\n',
                     'Valores válidos para "uf":\n', list(pvl_br['uf'].unique()))
  
````

````python
# 2) Seu código aqui
# Quantas solicitações estão registradas: tipo_interessado= Estado, uf= Minas Gerais, status= 'Arquivado por decurso de prazo'

pvl_mg_estado = monta_pvl_uf('MG','Estado')

# len(pvl_mg_estado[pvl_mg_estado['status']=='Arquivado por decurso de prazo'])

pvl_mg_estado['status'].value_counts()['Arquivado por decurso de prazo']
````

````python
# 3) Seu código aqui
# Qual é o município da Bahia com mais solicitações deferidas?

pvl_ba_mun = monta_pvl_uf('BA','Município')

deferidos_municipio_ba = pd.DataFrame(pvl_ba_mun[pvl_ba_mun['status']=='Deferido']['interessado'].value_counts())

deferidos_municipio_ba[deferidos_municipio_ba['count']==3]
````

````python
# 4) Seu código aqui

pvl_ba_estado = monta_pvl_uf('BA','Estado').to_csv('./Mod4_Ex03/pvl_ba_estado.csv')
````
