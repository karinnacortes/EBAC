## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
# Obs da aluna: Esse código já estava escrito no arquivo disponibilizado sem o index e com as colunas com o mesmo nome 'Qt_acertos'.
# Eu renomeei conforme achei melhor e coloquei as linhas de 1 a 100, pq de 0 a 99 me agoniza, rsrs. Tem problema? É uma má prática?

import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), index=range(1,101), columns=['Matematica'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), index=range(1,101), columns=['Portugues'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), index=range(1,101), columns=['Geografia'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), index=range(1,101), columns=['Ingles'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), index=range(1,101), columns=['Historia'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), index=range(1,101), columns=['Fisica'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), index=range(1,101), columns=['Quimica'])
````

````python
# Data frame com as variáveis disponibilizadas no código anterior.

vestibular_acertos = df_mat.join(df_por).join(df_geo).join(df_ing).join(df_his).join(df_fis).join(df_qui)
vestibular_acertos
````

````python
# 1) Seu código aqui
# Obs da aluna: Esse é meu código ;)

def percentual(acertos,total_questoes_materia):
    return round(acertos / total_questoes_materia,2)
pc_acertos_vestibular_materia = percentual(df_mat,24).join(percentual(df_por,18)).join(percentual(df_geo,8)).join(percentual(df_ing,8)).join(percentual(df_his,8)).join(percentual(df_fis,12)).join(percentual(df_qui,12))
pc_acertos_vestibular_materia    
````

````python
# 2) Seu código aqui

acertos_vestibular_total = vestibular_acertos.sum(axis=1)
pd.DataFrame(acertos_vestibular_total, columns=['Acertos no vestibular'])
````

````python
# 3) Seu código aqui

percentual(acertos_vestibular_total,90)
pd.DataFrame(percentual(acertos_vestibular_total,90),columns=['Percentual de acertos'])
````

````python
# 4) Seu código aqui

acertos_vestibular_total[acertos_vestibular_total > 45].count()
````

## 4) Vacinação II
Gere um *DataFrame* que contenha somente os estabelecimentos que aplicaram vcinas a menores de 18 anos. Nesse *DataFrame* devem conter somente os dados dos estabelecimentos, mais uma coluna sendo a quantidade de vacinas que o estabelecimento aplicou a menores de 18 anos.  
  
1. crie uma cópia do *DataFrame* original, contendo somente os registros de vacinas realizadas a menores de 18 anos.  
2. crie uma lista das colunas desse *DataFrame* com o atributo de *DataFrame* **.columns()**  
3. Nesse *DataFrame* faça uma contagem do campo ```vacina_categoria_nome```.
3. a partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma lista com esses valores.  
4. usando o método *.loc*, selecione somente essas variáveis  
5. Aplique o método **.drop_duplicates** e crie uma lista com uma linha para cada estabelecimento, com os dados do estabelecimento  

````python
# 1) 
# Registro dos menores de idade vacinados no Acre

df_menor = df[df["paciente_idade"]<18]
df_menor.reset_index(inplace=True, drop=True)
df_menor
````

````python
# 2) 
# lista dos nomes das colunas

list(df_menor.columns)
````

````python
# 3)

df_menor['vacina_categoria_nome'].value_counts()
````

````python
# 4)
# Informações sobre os estabelecimentos que atenderam cada menor.

colunas_info_estabelecimento = ['estabelecimento_valor',
                   'estabelecimento_razaosocial',
                   'estalecimento_nofantasia',
                   'estabelecimento_municipio_codigo',
                   'estabelecimento_municipio_nome',
                   'estabelecimento_uf'
                  ]
df_menor_info_estabelecimento = df_menor.loc[:,colunas_info_estabelecimento]
df_menor_info_estabelecimento.reset_index(inplace=True, drop=True)
df_menor_info_estabelecimento
````

````python
# 5)
# Lista de estabelecimentos que vacinaram menores de idade no Acre.

estabelecimento_vacinou_menor = df_menor_info_estabelecimento.drop_duplicates()
estabelecimento_vacinou_menor.reset_index(inplace=True, drop=True)
estabelecimento_vacinou_menor
````
